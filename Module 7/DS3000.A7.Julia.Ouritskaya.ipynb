## Question 5 (10 pts)

Calculate the Pearson Correlation coefficient of carat and price. Explain the results and discuss the strength of the correlation.

```python
correlation_coefficient, p_value = stats.pearsonr(diamonds_df['carat'], diamonds_df['price'])
print(f'The correlation coefficient is: {correlation_coefficient: .2f}')
print(f'The p-value is: {p_value: .2f}')
```

The result indicates that the correlation coefficient is 0.92, which is a strong positive correlation between carat (the weight of the diamond) and price. Thus, as the carat of the diamond increases, the price also increases. The p-value suggests that this correlation is statistically significant. It reveals that its unlikely that this pattern would appear if no correlation was evident.

## Question 7 (10 pts)

Using a 2-sample t-test**, determine if there is a statistical difference between the price of diamonds with a cut that is considered: (1) 'Good' vs 'Very Good' and (2) 'Premium' vs 'Ideal' (set alpha to 5% i.e. 0.05.)

```python
# Using a 2-sample t-test, determine if there is a statistical difference between the price of diamonds with a cut that is considered 'Good' vs. 'Very Good' (alpha to 5%)
good_cut = diamonds_df[diamonds_df['cut'] == 'Good']['price']
very_good_cut = diamonds_df[diamonds_df['cut'] == 'Very Good']['price']

alpha = 0.05
t_stat, p_value = stats.ttest_ind(good_cut, very_good_cut)

if p_value < alpha:
    print('There is a statistical difference between "Good" and "Very Good" cut prices.')
else:
        print('There is no statistical difference between "Good" and "Very Good" cut prices.')
```

The p-value is greater than the alpha level (0.05). Thus, we conclude that there is no statisitical difference between the prices of diamonds with `Good` and `Very Good` cuts. This means that, at a 5% significance level, we do not have enough evidence to claim that one cut is significantly more expensive than the other.

```python
# Using a 2-sample t-test, determine if there is a statistical difference between the price of diamonds with a cut that is considered 'Premium' vs. 'Ideal' (alpha to 5%)
premium_cut = diamonds_df[diamonds_df['cut'] == 'Premium']['price']
ideal_cut = diamonds_df[diamonds_df['cut'] == 'Ideal']['price']

alpha = 0.05
t_stat, p_value = stats.ttest_ind(premium_cut, ideal_cut)

if p_value < alpha:
    print('There is a statistical difference between "Premium" and "Ideal" cut prices.')
else:
        print('There is no statistical difference between "Premium" and "Ideal" cut prices.')
```

The p-value is less than the alpha level (0.05). Thus, we conclude that there is a statisitical difference between the prices of diamonds with `Premium` and `Ideal` cuts. This means that, at a 5% significance level, we have enough evidence to claim that one cut is significantly more expensive than the other.
