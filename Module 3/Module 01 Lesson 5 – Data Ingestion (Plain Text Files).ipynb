# Data Ingestion

## Plain Text Files
Python has a built-in function that allows us to read data from a text file. Its called the `open()` function. Let's use this function to read one of the works of William Shakespeare called _Richard III._ Credits: [The Folger Shakespeare](https://www.folger.edu/) 

```python
#Read the contents of a file
#the filename
file_name = 'RichardIII.txt' 
#r means read the file
fh = open(file_name, mode='r')
#start reading the contents of the file
data = fh.read()
fh.close()

print(type(data))
print(data)
```

```python
#Open the file using the context manager
with open(file_name, mode='r') as fh:
    data = fh.read()
    
print(data)
```

```python
#Open the file and read it sequentially
with open(file_name, mode='r') as fh:
    print(fh.readline())
    print(fh.readline())
    print(fh.readline())
```

```python
#Open the file and store its contents in a list
with open(file_name, mode='r') as fh:
    data = fh.readlines()
    
print(type(data))
```

```python
print(data[0])
print(data[1])
print(data[2])
```
